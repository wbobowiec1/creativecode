```
import random
import requests 
from textblob import TextBlob
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')




```

```

with open("/content/Wizard of Oz.txt") as newText:
  poemText = newText.read()

wordList = poemText.split(" ")

newBlob = TextBlob(poemText)
```

```
poem_text = """
My Sweet Miracle
 
When you finally came into this world
I knew right then, magic was real. 

The beautiful shades of your fur 
caught my eyes
Black. and White.
that's uniqueness

You stare at me with those eyes of yours 
Pitch black. 
Expressing the message of admiration
and Gentleness
This makes my heart skip couple of beats. 

I watch you lay peacefully here
dreams are taking over your mind 
While your adorable snores take over my ears 
your dreams flow through my mind too
I'll ensure a great future for them.

I wish you knew how much your smiles means 
your smile is a magnet which I'm jealous of 
for the power of attraction it has 
Everyday there are these array of colored eyes
They seem to not be able to take their eyes off of you
You seem to not mind that though, I'm glad. 
Your smiles makes all of our souls joyful. 

Others might think we need to diet 
But we don't care what they say
Our chubby bells is our pride and joy
We are lovers, not haters

Your daily routine is one i desire 
Go on your rocking horse for hours with 
dreams of actually riding one one day
then traveling to the field where it is 
Greener on the other side 
Rolling in the grass is your pleasure 
But not until you've eaten patches of it
Your life seems so playful and simple. 

"""

poemBlob = TextBlob(poem_text)
```

```
poemBlob.tags
```

```
singular_nouns = []
plural_nouns = []
adjectives = []

for word,pos in newBlob.tags:
  if(pos == 'NN'):
    singular_nouns.append(word)

  if(pos == 'NNS'):
    plural_nouns.append(word)
    
  if(pos == 'JJ'):
    adjectives.append(word)
```
